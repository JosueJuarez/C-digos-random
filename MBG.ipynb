Dado un proceso de Wiener $\{W(t) \,|\, 0 \geq t \geq T \}$ el movimiento browniano geométrico (MBG) es el proceso estcástico:
$$X(t) = X(0)exp\left\{ \left(\mu - \frac{1}{2} \sigma^{2} \right)t - \sigma W(t) \right\}, \quad X(0), \sigma > 0, \mu \in \mathbb{R}.$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dt = 1/255
N = 100
sigma = 0.3
mu = 0.8
X0 = 10.
```

```python
plt.figure(figsize=(20, 5))

n = 0
for i in range(0,10): #10 simulaciones
    
    norm = np.random.normal(0,1,N)
    X = np.exp((mu-((sigma**2)/2))*(1/255) + sigma*np.sqrt((1/255))*norm)
    X = np.insert(X, 0, X0)
    X = np.cumprod(X)
    
    plt.plot(X)
    
plt.title("Movimientos Brownianos Geométricos")
plt.show()
```
