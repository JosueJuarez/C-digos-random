Simulamos mediante el método de E-M, el MBG
$$ dX(t) = \mu X(t) dt + \sigma X(t) dW(t), \quad X(0) = x_0 \in \mathbb{R}^{+};$$
siendo $\mu, \sigma$ conocidos.
La solución es:
$$ X(t) = x_0 exp\left\{(\mu - \frac{1}{2} \sigma^{2}) t + \sigma W(t) \right\} $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 1.
n = 200
X0 = 10.
mu = 0.8
sigma = 0.3

t, dt = np.linspace(0,T,num=n,retstep=True)

W = np.random.normal(0,np.sqrt(dt), n-1) 
W = np.insert(W, 0, 0)
W = np.cumsum(W)

X = np.zeros(n)

X[0] = X0
```

```python
for i in range(1,n):
    X[i] = X[i-1] + mu*X[i-1]*dt + sigma*X[i-1]*(W[i] - W[i-1])
```

```python
plt.figure(figsize=(20, 5))
plt.plot(t,X, label="Euler-Maruyama")
plt.plot(t,X0*np.exp((mu-0.5*sigma**2)*t + sigma*W), "--", label="Solución de la EDE")
plt.title("E-M para el MBG")
plt.legend(loc='best')
plt.show()
```
