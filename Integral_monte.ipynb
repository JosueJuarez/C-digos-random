Aproximamos la integral de la función $g(x) = \log(x+1)/(x^2 +1)$ en el intervalo $[0,1]$:
$$I = \int_{0}^{1} \frac{\log(x+1)}{x^2 +1} dx = \frac{\pi}{8}\log(2),$$
utilizando las v.a. $X \sim U[0,1]$ cuyas f.d. son $f_{X}(x) = 1$.
Las v.a. $Y = g(X)/f_{X}(X)$ tiene $\mathbb{E}(Y) = I$.

Dadas las vv.aa. $X_1, X_2, \ldots$ i.i.d con densidad común $f_{X}$, por las ley fuerte de los grandes números
$$\frac{1}{n} \sum_{i}^{n} Y_{i} \rightarrow I  \text{ c.s.}$$

```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Integral_Approximation(n):
    x = np.random.rand(1,n) #n x_i v.a.
    y = np.log(x+1)/(x**2 + 1) #g(X)/f(X)
    approx = np.sum(y)/n

    return approx
```

```python
print(f"aproximación con n=100 es {Integral_Approximation(n=100)}")
print(f"I = {np.pi*np.log(2)/8}")
```

```python
approximations = [Integral_Approximation(n) for n in range(1, 10000)]
plt.plot(range(1, 10000), approximations)
plt.plot([1, 10000], [np.pi*np.log(2)/8, np.pi*np.log(2)/8], color="orange", label=r"$\frac{\pi}{8}\log(2)$")
plt.ylabel(r"Integral approximation")
plt.xlabel("n")
plt.legend(loc="best")
plt.show()
```
