Es el proceso $X(t)$ tal que
$$dX(t) = \frac{b - X(t)}{T-t} dt + dW(t), \quad X(0) = a, \quad b \in \mathbb{R}, \quad 0 \leq t \leq T.$$

La ecuación se resuelve usando la fórmula de Ito al proceso $Y(t) = X(t)/(T-t)$:
$$X(t) = \frac{a}{T} (T-t) + \frac{b}{T} t + (T-t) \int_{0}^{t} \frac{dW(s)}{T-s}.$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 100
n = 200
X0 = 10.
b = 10. #b = a

t, dt = np.linspace(0,T,num=n,retstep=True)

```

```python
plt.figure(figsize=(20, 5))

for i in range(0,5): #5 simulaciones

    W = np.random.normal(0,np.sqrt(dt), n-1) 
    W = np.insert(W, 0, 0)
    W = np.cumsum(W)

    X = np.zeros(n)

    X[0] = X0
    for i in range(1,n):
        X[i] = X[i-1] + ((b-X[i-1])/(T-t[i-1]))*dt + (W[i] - W[i-1])
        
    plt.plot(t,X)

plt.title("Simulaciones E-M para el Puente Browniano")
plt.show()
```
