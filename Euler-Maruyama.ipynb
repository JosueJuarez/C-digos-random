Si consideramos el siguiente proceso en forma diferencial
$$dX(t) = f(X(t))dt + g(X(t))dW(t), \quad X(0) = X_0, \quad 0 \leq t \leq T.$$

El método de Euler-Maruyama es el siguiente:
- Se discretiza el intervalo $[0,T]$  como $\delta t = T/N$, paral algún $N$.
- Consideramos $\tau_j = j\delta t$ , y $X(\tau_j) = X_j$.
- La solución se apróxima como
$$X_j = X_{j-1} + f(X_{j-1}) \delta t + g(X_{j-1}) ( W(\tau_j) - W(\tau_{j-1}) ), \quad j = 1,2,\ldots,N.$$

## Ejemplo

$$ dX(t) = 2 W(t) dW(t) + dt $$

La solución de esa ecuación es:
$$ X(t) = W^2 (t) + X_0,$$
con $X(0) = X_0$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 100.
n = 200
X0 = 5.

t, dt = np.linspace(0,T,num=n,retstep=True)

W = np.random.normal(0,np.sqrt(dt), n-1) 
W = np.insert(W, 0, 0)
W = np.cumsum(W)

X = np.zeros(n)

X[0] = X0
```

```python
for i in range(1,n):
    X[i] = X[i-1] + dt + 2*W[i-1]*(W[i] - W[i-1])
```

```python
plt.figure(figsize=(20, 5))
plt.plot(t,X,label="Aproximación E-M")
plt.plot(t,W*W + X0, "--", label="Solución de la EDE")
plt.legend(loc='best')
plt.show()
```
