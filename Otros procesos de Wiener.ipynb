Sea $\{W(t) \, : \, t \leq 0 \}$ un proceso de Wiener. Los siguientes pueden demotrarse que también son procesos de Wiener.

## 1. Reflexión
$B(t) = -W(t)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# calcular hasta X(T) en n pasos
T = 10
n = 1000
t, dt = np.linspace(0,T,num=n,retstep=True)
```

```python
W = np.random.normal(0,np.sqrt(dt), n-1) 
W = np.insert(W, 0, 0)
W = np.cumsum(W)
    
plt.figure(figsize=(20, 5))
plt.plot(t,W, label=r"$W(t)$")
plt.plot(t,-W, label=r"$B(t) = -W(t)$")
plt.title("Reflexión")
plt.legend(loc='best')
plt.show()
```

## 2. Escalamiento

$B(t) = cW(t/c^{2}), \, c \neq 0.$

```python
c = 3

W = np.random.normal(0,np.sqrt(dt)/c, n-1) 
W = np.insert(W, 0, 0)
W = np.cumsum(W)
    
plt.figure(figsize=(20, 5))
plt.plot((t)/(c**2),W, label=r"$W(t/c^{2})$")
plt.plot(t,c*W, label=r"$B(t) = cW(t/c^{2}), \, c \neq 0.$")
plt.title("Escalamiento")
plt.legend(loc='best')
plt.show()
```

## 3. Time inversion

$B(0) = 0$,

$B(t) = tW(1/t), \, t \neq 0.$

```python
def w(t): #t: float
    #n=1000
    dt = t/999 #dt = t/n-1
    Wtemp = np.random.normal(0,np.sqrt(dt), n-1) 
    
    return np.sum(Wtemp) #genera el elemento w(t) simulando n elementos antes de el

def W(t): #t: array
    Wtemp = np.zeros(n)
    for i in range(1,n):
        Wtemp[i] = w(t[i])
        
    return Wtemp #genera el vector W = (w1, ... , wn) con wi = w(ti) y ti son los elementos de un vector t
```

```python
plt.figure(figsize=(20, 5))
B = t*W(1/t)
plt.plot(t,B, label=r"$B(t) = tW(1/t), \, t \neq 0.$")
plt.title("Time inversion")
plt.legend(loc='best')
plt.show()
plt.show()

```

## 4. Time reversed

$B(t) = W(1) - W(1-t)$, 
$t \leq 1$

```python
plt.figure(figsize=(20, 5))

t2 = np.linspace(0,1,num=n)
w1 = w(1)
B = w1 - W(1-t2)

plt.plot(t2,B, label=r"$B(t) = W(1) - W(1-t), \, t \leq 1.$")
plt.title("Time inversion")
plt.legend(loc='best')
plt.show()
```

## 5. Time shifting

$B(t) = W(t+u) - W(u)$, $u < 0$.

```python
u = 10

plt.figure(figsize=(20, 5))

wu = w(u)
B = W(t+u) - wu

plt.plot(t,B, label=r"$B(t) = W(t+u) - W(u), \, u < 0.$")
plt.title("Time shifting")
plt.legend(loc='best')
plt.show()
```
